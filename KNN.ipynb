```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# 1ï¸ Veri Setini YÃ¼kleme
data = load_wine()
df = pd.DataFrame(data.data, columns=data.feature_names)
df['target'] = data.target

# 2ï¸ EÄŸitim ve Test Setlerine AyÄ±rma
X_train, X_test, y_train, y_test = train_test_split(df.iloc[:, :-1], df['target'], test_size=0.2, random_state=42)

# 3 Veriyi Ã–lÃ§eklendirme
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# 4 FarklÄ± Mesafe Ã–lÃ§Ã¼mleriyle KNN Modelleri
metrics = ['euclidean', 'manhattan', 'minkowski', 'hamming']
results = {}

for metric in metrics:
    if metric == 'hamming':
        X_train_bin = np.where(X_train > 0, 1, 0)  # Binary dÃ¶nÃ¼ÅŸÃ¼m
        X_test_bin = np.where(X_test > 0, 1, 0)
        knn = KNeighborsClassifier(n_neighbors=5, metric=metric)
        knn.fit(X_train_bin, y_train)
        y_pred = knn.predict(X_test_bin)
    else:
        knn = KNeighborsClassifier(n_neighbors=5, metric=metric)
        knn.fit(X_train, y_train)
        y_pred = knn.predict(X_test)

    acc = accuracy_score(y_test, y_pred)
    results[metric] = acc
    print(f"Mesafe Ã–lÃ§Ã¼mÃ¼: {metric} - DoÄŸruluk: {acc:.4f}")

# 5 SonuÃ§larÄ± GÃ¶rselleÅŸtirme
import matplotlib.pyplot as plt

plt.bar(results.keys(), results.values(), color=['blue', 'red', 'green', 'purple'])
plt.xlabel("Mesafe Ã–lÃ§Ã¼mÃ¼")
plt.ylabel("DoÄŸruluk Skoru")
plt.title("KNN FarklÄ± Mesafe Ã–lÃ§Ã¼mleri KarÅŸÄ±laÅŸtÄ±rmasÄ±")
plt.show()
```

```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import matplotlib.pyplot as plt
import seaborn as sns

# 1ï¸ Veri Setini YÃ¼kleme
data = load_wine()
df = pd.DataFrame(data.data, columns=data.feature_names)
df['target'] = data.target

# 2ï¸ EÄŸitim ve Test Setlerine AyÄ±rma
X_train, X_test, y_train, y_test = train_test_split(df.iloc[:, :-1], df['target'], test_size=0.2, random_state=42)

# 3ï¸ Veriyi Ã–lÃ§eklendirme
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# 4ï¸ KNN Modelini Uygulama
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)

# 5ï¸ DoÄŸruluk ve KarmaÅŸÄ±klÄ±k Matrisi
accuracy = accuracy_score(y_test, y_pred)
print(f"DoÄŸruluk: {accuracy:.4f}")

# KarmaÅŸÄ±klÄ±k Matrisi
conf_matrix = confusion_matrix(y_test, y_pred)

# Seaborn ile GÃ¶rselleÅŸtirme
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=data.target_names, yticklabels=data.target_names)
plt.xlabel('Tahmin Edilen SÄ±nÄ±f')
plt.ylabel('GerÃ§ek SÄ±nÄ±f')
plt.title('KarmaÅŸÄ±klÄ±k Matrisi')
plt.show()

# 6ï¸ SÄ±nÄ±flandÄ±rma Raporu
print("\nSÄ±nÄ±flandÄ±rma Raporu:")
print(classification_report(y_test, y_pred, target_names=data.target_names))

# 7ï¸ Karar SÄ±nÄ±rlarÄ±nÄ± GÃ¶rselleÅŸtirme (Sadece 2D Ã¶zellikler iÃ§in)
from sklearn.decomposition import PCA

# PCA ile 2D projeksiyon yapÄ±yoruz
pca = PCA(n_components=2)
X_train_pca = pca.fit_transform(X_train)
X_test_pca = pca.transform(X_test)

# KNN Modelini Uygulama (2D verilere)
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train_pca, y_train)
y_pred_pca = knn.predict(X_test_pca)

# Karar SÄ±nÄ±rlarÄ± GrafiÄŸi
plt.figure(figsize=(8, 6))
x_min, x_max = X_train_pca[:, 0].min() - 1, X_train_pca[:, 0].max() + 1
y_min, y_max = X_train_pca[:, 1].min() - 1, X_train_pca[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.02),
                     np.arange(y_min, y_max, 0.02))

Z = knn.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.contourf(xx, yy, Z, alpha=0.3)
plt.scatter(X_test_pca[:, 0], X_test_pca[:, 1], c=y_test, edgecolors='k', marker='o', s=50, cmap=plt.cm.coolwarm)
plt.title('KNN Karar SÄ±nÄ±rlarÄ± (PCA Ä°le 2D Projeksiyon)')
plt.xlabel('Ã–zellik 1 (PCA)')
plt.ylabel('Ã–zellik 2 (PCA)')
plt.show()
```
