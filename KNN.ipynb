# **K-En Yakın Komşu (KNN) Algoritması Nedir?**

KNN (K-Nearest Neighbors), makine öğrenmesinde sıkça kullanılan bir algoritmadır ve yeni bir veri noktasını sınıflandırmak veya tahmin yapmak için en yakın komşularıyla olan benzerliğini kullanır. Bu yöntem, veriyi öğrenmek yerine ezberleyen (instance-based learning) bir yöntem olduğu için "tembel öğrenme" (lazy learning) algoritması olarak da adlandırılır.

KNN’in temel varsayımı, benzer noktaların birbirine yakın olduğu fikrine dayanır. Bu, genellikle "benzer kuşlar birlikte uçar" (birds of a feather flock together) prensibiyle açıklanır.

##KNN iki farklı amaçla kullanılabilir:

1. Sınıflandırma (Classification): Yeni bir veri noktasını, komşularının çoğunluğunun ait olduğu sınıfa dahil eder.

2. Regresyon (Regression): Yeni bir veri noktasının değerini, en yakın komşularının ortalamasını alarak tahmin eder.

###KNN Algoritmasının Tarihçesi

🔹 1951 yılında Evelyn Fix ve Joseph Hodges tarafından ABD ordusu için yapılan araştırmalar kapsamında geliştirildi.

🔹 1967 yılında Thomas Cover ve Peter Hart, KNN ile ilgili ilk büyük çalışmaları yayımladı.

🔹 1980’lerde James Keller, hata oranlarını düşüren "bulanık KNN" (fuzzy KNN) modelini geliştirdi.

🔹 Günümüzde KNN, genetikten finans sektörüne kadar pek çok alanda yaygın olarak kullanılan bir algoritmadır.

##KNN Algoritması Nasıl Çalışır?

###Eğitim verileriyle algoritmayı oluşturur:

KNN, geleneksel makine öğrenme modelleri gibi bir öğrenme sürecinden geçmez. Bunun yerine, tüm eğitim verilerini hafızasında saklar.

###Yeni bir veri geldiğinde en yakın komşularını belirler:

Bir noktanın diğer veri noktalarına olan mesafesi hesaplanır.

###Sınıflandırma için çoğunluk oylaması yapılır:

En yakın komşuların çoğunluğu hangi sınıftaysa, yeni nokta da o sınıfa atanır.

###Regresyon için ortalama değeri hesaplar:

En yakın komşuların ortalama değeri alınarak tahmin yapılır.

##Oylama Türleri: Çoğunluk vs. Çoğulluk Oylaması

KNN’de sınıflandırma yaparken iki farklı oylama yöntemi kullanılır:

1. Çoğunluk Oylaması (Majority Voting): Eğer iki sınıf varsa, %50’den fazlası hangi sınıfa aitse, o sınıf seçilir.

2. Çoğulluk Oylaması (Plurality Voting): Eğer birden fazla sınıf varsa, en fazla oya sahip sınıf seçilir.

##Örnek - Çoğulluk Oylaması:

🔔🔔🔔💰💰💰🏠🏠🏠🏠

Çoğunluk Oylaması: Yok (hiçbir sınıf %50’den fazla değil)

Çoğulluk Oylaması: 🏠 (en fazla oy alan sınıf)

##KNN’de Kullanılan Mesafe Ölçümleri

KNN algoritmasının en önemli kısmı, veri noktaları arasındaki mesafenin hesaplanmasıdır. İşte en yaygın mesafe ölçümleri:

###Öklid Mesafesi (Euclidean Distance):

İki nokta arasındaki düz çizgi mesafesi olarak hesaplanır.
En yaygın kullanılan yöntemdir.

### Manhattan Mesafesi (Taxicab Distance):

Sadece yatay ve dikey hareketleri hesaplar.
Taksi mesafesi olarak da bilinir.

###Minkowski Mesafesi:

Öklid ve Manhattan mesafelerinin genelleştirilmiş hali olarak düşünebiliriz.
Eğer p=1 ise Manhattan mesafesi, p=2 ise Öklid mesafesi kullanılır.

###Hamming Mesafesi:

İkili (binary) veya metin verilerinde farklı karakterleri saymak için kullanılır.

Örneğin: "10110" ve "10011" dizilerinde farklı olan bitleri sayarak mesafe ölçer.Bu iki dizede 2 farklı bit var o yüzden hamming mesafesi ikidir.

##K Değerini Belirleme

KNN’nin başarısı seçilen K değeriyle doğrudan ilişkilidir.

1. K değeri çok küçükse: Model aşırı öğrenme (overfitting) yapar, yani eğitim verisine fazla bağımlı olur.

2. K değeri çok büyükse: Model genelleme yapamaz ve az öğrenme (underfitting) problemi yaşanır.

###İdeal K değeri nasıl seçilir?

Genellikle K = √N (N, veri sayısı) bir başlangıç noktası olarak kullanılır.

Çapraz doğrulama (cross-validation) yöntemleriyle en iyi K değeri belirlenir.

##KNN Algoritmasının Avantajları

Basit ve anlaşılır: KNN, en kolay uygulanabilen makine öğrenme algoritmalarından biridir.

Eğitim süresi yok: Modeli oluştururken uzun süren bir eğitim aşaması gerektirmez.

Yeni veriye kolay adapte olur: Yeni veri eklendiğinde modeli baştan eğitmeye gerek yoktur.

##KNN Algoritmasının Zorlukları ve Dezavantajları

Hafıza kullanımı fazla: KNN, bütün eğitim verisini sakladığı için fazla yer kaplar.

Boyutlar arttıkça başarısı düşer (Curse of Dimensionality):

Veri boyutu arttıkça mesafeler arasındaki farklar küçülür ve benzer noktaları ayırt etmek zorlaşır.

Öğrenme süreci yavaştır: Eğitim süresi olmamasına rağmen, tahmin yaparken tüm veri setini taramak gerektiğinden yavaştır.

##KNN Algoritmasının Kullanım Alanları


**Doğal Dil İşleme (NLP):** Arama motorlarında en alakalı sonuçları sıralamak için kullanılır.

**Görüntü Tanıma:** Benzer resimleri ve videoları bulmak için kullanılır.

**Finans:** Hisse senedi fiyat tahmini, döviz kurları analizi gibi finansal tahminlerde kullanılır.

**Ürün Öneri Sistemleri:** Netflix, Amazon gibi platformlar, KNN kullanarak kişisel öneriler sunar.

**Sağlık:** Genetik analizlerde, hastalık tahminlerinde ve teşhislerde kullanılır.

**Eksik Veri Tamamlama:** Eksik verileri tahmin etmek için KNN kullanılabilir.

```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# 1️ Veri Setini Yükleme
data = load_wine()
df = pd.DataFrame(data.data, columns=data.feature_names)
df['target'] = data.target

# 2️ Eğitim ve Test Setlerine Ayırma
X_train, X_test, y_train, y_test = train_test_split(df.iloc[:, :-1], df['target'], test_size=0.2, random_state=42)

# 3 Veriyi Ölçeklendirme
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# 4 Farklı Mesafe Ölçümleriyle KNN Modelleri
metrics = ['euclidean', 'manhattan', 'minkowski', 'hamming']
results = {}

for metric in metrics:
    if metric == 'hamming':
        X_train_bin = np.where(X_train > 0, 1, 0)  # Binary dönüşüm
        X_test_bin = np.where(X_test > 0, 1, 0)
        knn = KNeighborsClassifier(n_neighbors=5, metric=metric)
        knn.fit(X_train_bin, y_train)
        y_pred = knn.predict(X_test_bin)
    else:
        knn = KNeighborsClassifier(n_neighbors=5, metric=metric)
        knn.fit(X_train, y_train)
        y_pred = knn.predict(X_test)

    acc = accuracy_score(y_test, y_pred)
    results[metric] = acc
    print(f"Mesafe Ölçümü: {metric} - Doğruluk: {acc:.4f}")

# 5 Sonuçları Görselleştirme
import matplotlib.pyplot as plt

plt.bar(results.keys(), results.values(), color=['blue', 'red', 'green', 'purple'])
plt.xlabel("Mesafe Ölçümü")
plt.ylabel("Doğruluk Skoru")
plt.title("KNN Farklı Mesafe Ölçümleri Karşılaştırması")
plt.show()
```

```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import matplotlib.pyplot as plt
import seaborn as sns

# 1️ Veri Setini Yükleme
data = load_wine()
df = pd.DataFrame(data.data, columns=data.feature_names)
df['target'] = data.target

# 2️ Eğitim ve Test Setlerine Ayırma
X_train, X_test, y_train, y_test = train_test_split(df.iloc[:, :-1], df['target'], test_size=0.2, random_state=42)

# 3️ Veriyi Ölçeklendirme
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# 4️ KNN Modelini Uygulama
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)

# 5️ Doğruluk ve Karmaşıklık Matrisi
accuracy = accuracy_score(y_test, y_pred)
print(f"Doğruluk: {accuracy:.4f}")

# Karmaşıklık Matrisi
conf_matrix = confusion_matrix(y_test, y_pred)

# Seaborn ile Görselleştirme
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=data.target_names, yticklabels=data.target_names)
plt.xlabel('Tahmin Edilen Sınıf')
plt.ylabel('Gerçek Sınıf')
plt.title('Karmaşıklık Matrisi')
plt.show()

# 6️ Sınıflandırma Raporu
print("\nSınıflandırma Raporu:")
print(classification_report(y_test, y_pred, target_names=data.target_names))

# 7️ Karar Sınırlarını Görselleştirme (Sadece 2D özellikler için)
from sklearn.decomposition import PCA

# PCA ile 2D projeksiyon yapıyoruz
pca = PCA(n_components=2)
X_train_pca = pca.fit_transform(X_train)
X_test_pca = pca.transform(X_test)

# KNN Modelini Uygulama (2D verilere)
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train_pca, y_train)
y_pred_pca = knn.predict(X_test_pca)

# Karar Sınırları Grafiği
plt.figure(figsize=(8, 6))
x_min, x_max = X_train_pca[:, 0].min() - 1, X_train_pca[:, 0].max() + 1
y_min, y_max = X_train_pca[:, 1].min() - 1, X_train_pca[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.02),
                     np.arange(y_min, y_max, 0.02))

Z = knn.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.contourf(xx, yy, Z, alpha=0.3)
plt.scatter(X_test_pca[:, 0], X_test_pca[:, 1], c=y_test, edgecolors='k', marker='o', s=50, cmap=plt.cm.coolwarm)
plt.title('KNN Karar Sınırları (PCA İle 2D Projeksiyon)')
plt.xlabel('Özellik 1 (PCA)')
plt.ylabel('Özellik 2 (PCA)')
plt.show()
```
