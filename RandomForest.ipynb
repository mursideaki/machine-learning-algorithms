# **RANDOM FOREST**

**Random Forest Algoritması Nedir?**

Random Forest, bir ensemble öğrenme yöntemidir ve hem sınıflandırma hem de regresyon problemleri için kullanılır. Bu algoritma, Karar Ağaçlarının bir koleksiyonunu (forest) oluşturur ve bu ağaçların sonuçlarını birleştirerek nihai tahmini yapar.

**Ana Fikir Nedir?**

**Bagging (Bootstrap Aggregation)**

Veri setinden rastgele örnekler seçilerek birden fazla karar ağacı eğitilir. Bu, bireysel modellerin çeşitliliğini artırır. Her ağaç, farklı bir veri alt kümesine göre eğitildiğinden, her biri bağımsızdır.

**Örnekleme Yöntemi:**
Orijinal veri setinin boyutunda rastgele örnekler alınır (tekrarlı seçime izin verilir).
Örneklerin yaklaşık %63'ü bir ağaç için eğitime dahil edilir (out-of-bag (OOB) örnekler ise değerlendirme için kullanılabilir).


**Rastgele Özellik Seçimi**

Her bir karar ağacı oluşturulurken, bölünmeler için kullanılan özellikler rastgele seçilir. Bu, ağaçlar arasındaki korelasyonu azaltır ve modelin daha iyi genelleştirme yapmasını sağlar.

**Nasıl Çalışır?**

**Eğitim Aşaması**

Veri setinden rastgele örnekler alınır (bootstrap örnekleme).
Her örnekleme setiyle bir karar ağacı eğitilir.
Her düğümde, bölünme için sadece rastgele seçilen bir alt özellik kümesi değerlendirilir.

**Tahmin Aşaması**

**Sınıflandırma için:**

Her bir ağacın tahmini alınır ve çoğunluk oyu kullanılarak sınıf belirlenir.

**Regresyon için:**

Her bir ağacın tahmini alınır ve ortalaması hesaplanır.

# Random Forest Ne Zaman Kullanılmalı?

**Büyük ve gürültülü veri kümelerinde:**

Random Forest, veri gürültüsüne karşı dayanıklıdır ve aşırı uyuma (overfitting) daha az eğilim gösterir.

**Yorumlanabilirliğin önemli olduğu durumlarda:**

Random Forest, bireysel ağaçlar sayesinde özelliklerin model üzerindeki etkisini anlamayı kolaylaştırır.

**Daha hızlı eğitim süreleri gerektiğinde:**

Paralel bir şekilde ağaç oluşturabilen random forest algoritması, sıralı gradient boosting'e göre daha hızlı eğitilir.

**Sınırlı veriyle çalışırken:**

Küçük veri kümeleriyle bile rastgele ormanlar iyi performans gösterebilir.



# **Hyperparametre Ayarları**
Random Forest'ın performansını optimize etmek için bazı önemli hiperparametreler vardır:

**n_estimators:** Eğitilecek ağaç sayısı. Varsayılan olarak 100 ağaç ayarlanmıştır. Daha fazla ağaç genellikle daha iyi performans sağlar, ancak işlem süresini artırır.

**max_features:** Diyelim ki veri setimizdeki toplam özellik sayısı 100 ise ve bu özelliklerin karekökünü rastgele seçmeye karar verirsem, bu sayı 10 olur. Bu durumda, her düğümde rastgele 10 özellik seçeceğim ve en iyi bölmeyi yapmak için bu özellikler arasından seçim yapacağım. Diğer düğümler için de benzer şekilde, her seferinde rastgele 10 özellik seçeceğim.

Alternatif olarak, karekök almak yerine veri setindeki toplam özellik sayısının logaritmasını almayı da tercih edebilirim. Bu durumda, 'max_features' parametresi, her düğümde rastgele seçilecek özellik sayısını belirlemenize olanak tanır.

*   auto veya sqrt: Sınıflandırmada, özelliklerin karekökü kadar. Varsayılan olarak sqrt ayarlanmıştır.

*   log2: Özelliklerin logaritmik tabanı kadar.

Ya da kendim bir sayı da seçebilirim mesela max_features=5 gibi.

**max_depth:** Her bir ağacın maksimum derinliğini ifade eder. Yani max_depth, her ağacın alabileceği maksimum bölünme sayısını belirler. max_depth çok düşükse, model daha az eğitilir ve yüksek bir önyargıya sahip olur, bu da modelin yetersiz öğrenmesine (underfitting) yol açar. Aynı şekilde, max_depth yüksekse, model ezberlemeye yönelir ve uyumsuz bir model elde ederiz. Overfitting’i önlemek için max_depth sınırlandırılabilir, varsayılan olarak "None" ayarlanmıştır.

**min_samples_split:** Herhangi bir düğümdeki karar ağacı gözlemlerinin minimum sayısını belirler ve ona göre böler. Varsayılan olarak 2 ayarlanmıştır. Her düğümün iki alt düğümü olduğunu anlamına gelir.

**min_samples_leaf:** Bir yaprak düğümde bulunması gereken minimum örnek sayısı. varsayıklan olarak 1 ayarlanmıştır. Yüksek sayılarda ayarlanması örneklemenin düzgün yapılmasına engel olabilir.

**bootstrap:** Bagging yöntemi kullanılıp kullanılmayacağı ile ilgili parametredir. Varsayılan olarak "True" ayarlanmıştır. Aşağıya Random Forest için tüm parametre ayarlarını görebileceğiniz görsel eklenmiştir.

**Out-of-Bag (OOB) Error**

Bagging sırasında her ağaç, eğitim sırasında veri setinin yalnızca bir kısmını kullanır. Kullanılmayan OOB örnekler, modelin doğruluğunu değerlendirmek için kullanılabilir.
Bu, ayrı bir test seti ayırmadan modelin genelleme hatasını ölçmeye olanak tanır. Varsayılan değeri "False" olarak ayarlanmıştır.

**Özellik Önem Skoru**

Random Forest, her bir özelliğin model performansına olan katkısını ölçmek için bir özellik önem skoru hesaplar.

Gini Impurity veya Information Gain kullanılarak hesaplanır.
Daha yüksek önem skoru, özelliğin model tahminleri için daha önemli olduğunu gösterir.

# **Modelin Avantajları ve Kısıtlamaları**

**Avantajlar**

**Aşırı Uyumdan (Overfitting) Kaçınma:** Bagging ve rastgele özellik seçimi, overfitting'i büyük ölçüde azaltır.

**Esneklik:** Hem sınıflandırma hem de regresyon problemlerinde etkili bir şekilde kullanılabilir.

**Özellik Önem Skoru:** Model, hangi özelliklerin tahminler için daha önemli olduğunu belirleyebilir.

Karmaşık veri setlerinde bile yüksek doğruluk sağlar.

Kategorik ve sürekli verilerle kolayca çalışabilir.

**Dezavantajları**

Çok fazla ağaç, modelin eğitim ve tahmin sürelerini uzatabilir.

Büyük veri setlerinde, bellek kullanımı bir sorun olabilir.

Daha fazla hesaplama gücü gerektirir, çünkü birden fazla karar ağacı eğitilir.

Yorumlanabilirlik, tek bir karar ağacına göre daha düşüktür.


# **Random Forest’ın Diğer Algoritmalarla Karşılaştırılması**

# **Decision Tree**

**Basit ve anlaşılır bir model gerektiğinde:**

Eğer modelin kolayca yorumlanabilir olması önemliyse, karar ağacı kullanmayı tercih edin. Karar ağaçları, sonuçları anlamayı ve açıklamayı kolaylaştırır.

**Karmaşık ve büyük veri kümelerinde:**

Eğer veri kümeniz yüksek boyutlu özellikler içeriyor ve bu özelliklerle hedef değişken arasında karmaşık ilişkiler varsa, rastgele orman genelde daha iyi sonuç verir. Rastgele ormanlar, aşırı uyuma karşı dayanıklıdır ve genellikle daha yüksek doğruluk sunar.

**Küçük ve basit veri kümelerinde:**
Veri kümeniz küçükse ve hesaplama süresi sizin için kritikse, karar ağaçları daha hızlı ve uygun bir seçenek olabilir.


**Daha iyi genelleme performansı için:**
Büyük ve karmaşık veri kümeleri üzerinde çalışıyorsanız, rastgele ormanlar daha iyi genelleme yapabilir ve hataları minimize edebilir.

# **Gradient Boosting**

**Yüksek doğruluk gerektiğinde:**

Gradient boosting, özellikle verilerdeki karmaşık ilişkileri öğrenmek için tasarlanmıştır ve genellikle yüksek doğruluk sağlar.

**Küçük ve temiz veri kümelerinde:**

Verileriniz temiz ve düzenliyse, gradient boosting overfitting riskine karşı daha iyi bir seçenektir.

**Yorumlanabilirlik öncelikli olmadığında:**

Gradient boosting modelleri, random forest algoritmasına kıyasla daha az yorumlanabilir olabilir.

**Esnek kayıp fonksiyonları gerektiğinde:**

Gradient boosting, eğitim sırasında optimize edilecek kayıp fonksiyonunu özelleştirme esnekliği sunar. Bu, karmaşık problemler üzerinde avantaj sağlayabilir.

Random Forest, eğitim sırasında hata oranını düşürmeye odaklanmaz, bu yüzden genellikle daha hızlıdır. Ancak Gradient Boosting, daha yüksek doğruluk sağlayabilir.

# **Support Vector Machines**

**Random Forest ve SVM Arasında Seçim Yapma**

Random Forest ve Destek Vektör Makineleri (SVM), farklı avantaj ve dezavantajlara sahiptir. Hangi algoritmanın kullanılacağına karar verirken veri kümenizin özellikleri ve problem tanımınız önemlidir.

**1. Veri Kümesi Boyutu ve Karmaşıklığı**

**Random Forest:** Büyük ve yüksek boyutlu (çok sayıda özelliğe sahip) veri kümelerinde iyi performans gösterir. Özellik rastgeleliği sayesinde, büyük miktarda veriyi etkili bir şekilde işleyebilir.

**SVM:** Küçük veya orta boyutlu, düşük boyutlu ve iyi yapılandırılmış veri kümelerinde daha iyi çalışır.

**2. Veri Türü**

**Random Forest:** Karmaşık ve doğrusal olmayan desenleri kolayca yakalayabilir ve özellikler arasındaki ilişkileri öğrenebilir.

**SVM:** Eğer sınıflar doğrusal olarak ayrılabiliyorsa, SVM mükemmel bir sonuç verir. Ancak, kernel yöntemleri sayesinde doğrusal olmayan veri kümelerinde de etkili olabilir.

**3. Hesaplama Verimliliği**

**Random Forest:** Karar ağaçlarını paralel bir şekilde eğitebildiği için hesaplama açısından daha verimlidir.

**4. Marj (Sınıflar Arasındaki Mesafe) Önemliyse**

**SVM:** Maksimum marj optimizasyonu yaparak net ve güçlü bir karar sınırı oluşturur. Eğer sınıflar arasında belirgin bir ayrım gerekiyorsa, SVM uygun bir seçimdir.

**5. Özellik Önemi (Feature Importance)**

**Random Forest:** Her özelliğin göreceli önemini sıralayarak, veri kümenizdeki özelliklerin etkisini anlamanıza yardımcı olur.

**6. Yorumlanabilirlik**

**Random Forest:** Genel olarak daha yorumlanabilir bir modeldir.

**SVM:** Daha belirgin karar sınırları sayesinde, bazı uygulamalarda yorumlanabilirliği daha faydalı olabilir.

**7. Hiperparametre Ayarlama**

**Random Forest:** Hiperparametre ayarı gereksinimi genelde daha azdır ve kullanıcı dostudur.

**SVM:** Daha fazla hiperparametre ayarlaması gerektirebilir, bu da deneyimli kullanıcılar için daha uygun olabilir.

**8. Eğitim Süresi**

Eğer eğitim süresi önemli bir kriterse, Random Forest, paralel hesaplama sayesinde büyük veri kümelerinde daha hızlı eğitilebilir.

**9. Tekil Model vs. Topluluk Yaklaşımı**

**SVM:** Tek bir modeldir.

**Random Forest:** Bir karar ağacı topluluğudur (ensemble). Topluluk stratejisi kullanmak, model performansını artırabilir.

Hangi algoritmayı seçeceğinize karar verirken veri kümenizin boyutu, yapısı, modelin yorumlanabilirliği ve eğitim süresi gibi kriterleri göz önünde bulundurmalısınız. Random Forest, daha karmaşık ve büyük veri kümelerinde etkili olurken, SVM daha küçük ve yapılandırılmış veri kümelerinde mükemmel sonuçlar verebilir.

```python
# Gerekli kütüphaneler
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report

# Veri setini yükle
wine = load_wine()
X = wine.data
y = wine.target

# Eğitim ve test setlerine ayır
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Random Forest Modeli
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)

# Tahmin yap
y_pred = rf_model.predict(X_test)

# Performans Değerlendirme
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.2f}")
print("Classification Report:")
print(classification_report(y_test, y_pred))
```

# **Model Değerlendirme Metrikleri**
Bir modelin başarısını değerlendirirken çeşitli metrikler kullanılır. Her biri farklı bir bakış açısı sunar ve problemin gereksinimlerine göre farklı önem taşır.

# Accuracy (Doğruluk)
Modelin doğru tahmin ettiği örneklerin toplam örnek sayısına oranıdır.

**Avantaj:** Genel bir başarı ölçütüdür.

**Dezavantaj:** Dengesiz veri setlerinde yanıltıcı olabilir. Örneğin, %90'ı bir sınıfa ait bir veri setinde, hep aynı sınıfı tahmin eden bir model yüksek doğruluk sağlayabilir ancak gerçekte kötü bir performansa sahiptir.

# Precision (Kesinlik)
Bir sınıf için yapılan pozitif tahminlerin ne kadarının doğru olduğunu gösterir. Özellikle yanlış pozitifleri (false positive) minimize etmek istiyorsanız önemlidir.

**Kullanım:** E-posta spam filtresi gibi yanlış pozitiflerin maliyetli olduğu durumlarda karar vermenizi sağlayacak metrik kesinlik olmalıdır.

# Recall (Duyarlılık / Hassasiyet)
Modelin, pozitif sınıfları ne kadar iyi yakaladığını gösterir. Yanlış negatifleri (false negative) minimize etmek istiyorsanız önemlidir.


**Kullanım:** Hastalık teşhisi gibi yanlış negatiflerin daha önemli olduğu durumlarda recall metriğine bakmak önem taşır.

# F1-Score
Precision ve recall'un harmonik ortalamasıdır. Dengeli bir değerlendirme sağlar ve iki metriği birleştirir.

**Kullanım:** Precision ve recall arasında denge aradığınız durumlarda kullanımı doğru tercih olacaktır ve genel olarak modelinizin başarısını en iyi yansıtan metrik F1 Skorudur.

Aşağıdaki görselde bir karmaşıklık matrisi görmektesiniz. Karmaşıklık matrisi (Confusion Matrix) modelimizin başarısını net bir şekilde görmemizi sağlar.

Karmaşıklık matrisi, modelin yaptığı tüm tahminleri ve gerçek sınıfları gösteren bir tablodur. Bu tablo, modelin hangi sınıfları karıştırdığını ve hangi tür hatalar yaptığını görmemizi sağlar.

**True Positive (TP):** Doğru şekilde pozitif tahmin edilenler.

**False Positive (FP):** Yanlış şekilde pozitif tahmin edilenler (yani gerçekte negatif olanlar).

**False Negative (FN):** Yanlış şekilde negatif tahmin edilenler (yani gerçekte pozitif olanlar).

**True Negative (TN):** Doğru şekilde negatif tahmin edilenler.

**Confusion Matrix'in Yorumlanması**

Sınıflandırma hatalarının nereden kaynaklandığını anlamanızı sağlar. Yani model hangi etiketleri karıştırıyor bu şekilde anlayabilirsiniz.

Precision, recall ve diğer metrikler confusion matrix'ten hesaplanabilir.

**Dengesiz veri setlerinde**, doğruluk metrikine güvenmek yanıltıcı olabilir. Örneğin, %90'ı bir sınıfa ait olan bir veri setinde, her zaman çoğunluk sınıfını tahmin eden bir model yüksek doğruluk elde edebilir, ancak aslında kötü bir performans sergileyebilir. Bu gibi durumlarda, precision, recall, F1-score, ROC eğrisi ve AUC gibi metrikler kullanmak daha uygun olacaktır.

```python
# Gerekli kütüphaneler
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

# Confusion Matrix'i hesapla
cm = confusion_matrix(y_test, y_pred, labels=rf_model.classes_)

# Görselleştir
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=wine.target_names)
disp.plot(cmap=plt.cm.Blues)
plt.title("Confusion Matrix")
plt.show()
```

Yukarıda gördüğümüz confusion matrix şöyle yorumlanır; solda gördüğümüz eksen sınıflarımızın test datasındaki gerçek etiketleri aşağıda gördüğümüz eksense modelimizin tahminlerini gösterecektir. Mavi renkli köşegenin içerisinde bulunan sayılar kaç adet etiketin doğru tahmin edildiğini gösterirken bu köşegenin diğer iki tarafında kalan kısımlardaki sayılar yanlış tahminlerimizi gösterecektir. Bu eğitimde accuracy 1 olduğundan bütün örnekler doğru tahmin edilmiştir o nedenle köşegen dışında sayı görmüyorsunuz.

Aşağıda önem skoru ile ilgili kod parçası ve çıktısı görüyorsunuz. Bu bazı öznitelikleri eleyip o şekilde eğitim yapmanızı sağlayabileceği gibi isterseniz bu şekilde öz niteliklerinizi seçip başka algoritmada kullanabilirsiniz.

```python
import pandas as pd
import seaborn as sns

# Özellik önemlerini al
feature_importances = rf_model.feature_importances_
features = wine.feature_names

# Görselleştir
importance_df = pd.DataFrame({'Feature': features, 'Importance': feature_importances})
importance_df = importance_df.sort_values(by='Importance', ascending=False)

sns.barplot(x='Importance', y='Feature', data=importance_df)
plt.title("Feature Importances")
plt.show()
```
